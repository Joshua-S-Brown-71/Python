```python
from pprint import pprint
dic={}
count=0
r = open("//Users/JOSH/Desktop/Math 4332            (Python)/3. HW/HW9/HW9txtFiles/grades.txt")
for line in r:
    ls=line.split()
    dic[ls[0]]=[ls[1],ls[2],ls[3]]  
pprint(dic)
print("\n")
for s in dic.values():
    flag=1
    for i in s:
        if int(i)<70:
            flag=0
    if flag==1:
        count+=1
print("Number of students who passed all three tests: ",count)
r.close()
```

```python
from pprint import pprint
from string import punctuation

dic={}
dic3={}
l=[]
r = open("//Users/JOSH/Desktop/Math 4332            (Python)/3. HW/HW9/HW9txtFiles/wordlist.txt")
for line in r:
    dic2={}
    oline=line
    line=line.strip()
    line=line.lower()
    for char in punctuation:
        line=line.replace(char,"")
    for c in line:
        dic2[c]=line.count(c)
    dic[oline]=dic2
    
word = input("Enter a string: ").lower()
for c in word:
  if c not in dic3:
    dic3[c]=word.count(c)
print("\n")

for k,v in dic.items():
    if v==dic3:
        print(k)
```
